```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.pdfgen import canvas
import calendar


# Read the test data file into a pandas DataFrame
df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/review_dataset.csv")


def analyze_reviews():
    # Calculate the count of different review ratings
    review_counts = df['stars'].value_counts()

    # Plotting the reviews using a bar plot
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x='stars', data=df)
    plt.xlabel('Review Rating')
    plt.ylabel('Count')
    plt.title('Distribution of Reviews by Rating')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('reviews_analysis.png')  # Save the chart as an image
    plt.show()

    # Generate PDF report
    pdf_file = 'reviews_analysis_report.pdf'
    c = canvas.Canvas(pdf_file)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(220, 800, 'Analysis Report - Customer Reviews')
    c.drawImage('reviews_analysis.png', 50, 550, width=500, height=300)
    c.save()

    # Generate Excel file
    excel_file = 'reviews_analysis_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        review_counts.to_excel(writer, sheet_name='Reviews Analysis')
        
        
# Option 2: Analysis of different payment methods used by the Customers
def analyze_payment_methods():
    
    df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/orders_2016-2020_Dataset.csv")

    # Analysis of different payment methods used by Customers
    payment_counts = df['Payment Method'].value_counts()
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Payment Method', data=df)
    plt.xlabel('Payment Method')
    plt.ylabel('Count')
    plt.title('Analysis of different payment methods used by Customers')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('payment_methods_analysis.png')  # Save the chart as an image
    plt.show()

    # Generate PDF report
    pdf_file = 'payment_methods_analysis_report.pdf'
    c = canvas.Canvas(pdf_file)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(200, 800, 'Analysis Report - Payment Methods')
    c.drawImage('payment_methods_analysis.png', 50, 550, width=500, height=300)
    c.save()

    # Generate Excel file
    excel_file = 'payment_methods_analysis_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        payment_counts.to_excel(writer, sheet_name='Payment Methods Analysis')

    # Option 3: Analysis of Top Consumer States of India
def analyze_consumer_states():
    df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/orders_2016-2020_Dataset.csv")
    # Add your code for consumer states analysis here
    
    # Calculate the count of orders by state
    state_counts = df['Shipping State'].value_counts().head(10)

    # Plotting the top consumer states using a bar plot
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=state_counts.index, y=state_counts.values)
    plt.xlabel('Consumer State')
    plt.ylabel('Count')
    plt.title('Top Consumer States of India')
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Save the bar plot as an image
    plt.savefig('top_consumer_states.png')
    
    # Show the bar plot
    plt.show()
    
    # Generate PDF report
    pdf_file = 'top_consumer_states_report.pdf'
    c = canvas.Canvas(pdf_file)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(200, 800, 'Analysis Report - Top Consumer States')
    c.drawImage('top_consumer_states.png', 50, 550, width=500, height=300)
    c.save()

    # Generate Excel file
    excel_file = 'top_consumer_states_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        state_counts.to_excel(writer, sheet_name='Top Consumer States')
 #option 4       
def analyze_consumer_cities():
    df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/orders_2016-2020_Dataset.csv")
    # Calculate the count of orders by city
    city_counts = df['Shipping City'].value_counts().head(10)

    # Plotting the top consumer cities using a bar plot
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=city_counts.index, y=city_counts.values)
    plt.xlabel('Consumer City')
    plt.ylabel('Count')
    plt.title('Top Consumer Cities of India')
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Save the bar plot as an image
    plt.savefig('top_consumer_cities.png')
    
    # Show the bar plot
    plt.show()

    # Generate PDF report
    pdf_file = 'top_consumer_cities_report.pdf'
    c = canvas.Canvas(pdf_file)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(180, 800, 'Analysis Report - Top Consumer Cities')
    c.drawImage('top_consumer_cities.png', 50, 550, width=500, height=300)
    c.save()

    # Generate Excel file
    excel_file = 'top_consumer_cities_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        city_counts.to_excel(writer, sheet_name='Top Consumer Cities')
  

 #option 5     
def analyze_product_categories():
    df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/review_dataset.csv")

    category_counts = df['category'].value_counts().head(10)

    # Plotting the top selling product categories using a bar plot
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=category_counts.index, y=category_counts.values)
    plt.xlabel('Product Category')
    plt.ylabel('Count')
    plt.title('Top Selling Product Categories')
    plt.xticks(rotation=45)
    plt.tight_layout()

   
    plt.savefig('top_product_categories.png')

    # Show the bar plot
    plt.show()

    # Generate PDF report
    pdf_file = 'top_product_categories_report.pdf'
    c = canvas.Canvas(pdf_file)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(150, 800, 'Analysis Report - Top Selling Product Categories')
    c.drawImage('top_product_categories.png', 50, 550, width=500, height=300)
    c.save()

    # Generate Excel file
    excel_file = 'top_product_categories_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        category_counts.to_excel(writer, sheet_name='Top Selling Product Categories')

#option 6
def analyze_reviews_product_categories():
    df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/review_dataset.csv")
    df['stars'] = df['stars'].str.extract('(\d+\.?\d*)').astype(float)

    category_reviews = df.groupby('category')['stars'].mean()

    plt.figure(figsize=(20, 6))
    ax = sns.barplot(x=category_reviews.index, y=category_reviews.values)
    plt.xlabel('Product Category')
    plt.ylabel('Average Review Rating')
    plt.title('Reviews for All Product Categories')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Save the bar plot as an image
    plt.savefig('category_reviews.png')

    # Show the bar plot
    plt.show()

    # Generate PDF report
    pdf_file = 'category_reviews_report.pdf'
    c = canvas.Canvas(pdf_file)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(150, 800, 'Analysis Report - Reviews for All Product Categories')
    c.drawImage('category_reviews.png', 50, 550, width=800, height=400)
    c.save()

    # Generate Excel file
    excel_file = 'category_reviews_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        category_reviews.to_excel(writer, sheet_name='Reviews for Product Categories')


#option 7

from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter

def analyze_orders_per_month():
    df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/orders_2016-2020_Dataset.csv")
    
    # Convert 'Order Date and Time Stamp' column to datetime
    df['Order Date and Time Stamp'] = pd.to_datetime(df['Order Date and Time Stamp'], format='%d-%m-%Y %H:%M:%S %z')

    # Extract year and month from the 'Order Date and Time Stamp' column
    df['Year'] = df['Order Date and Time Stamp'].dt.year
    df['Month'] = df['Order Date and Time Stamp'].dt.month

    # Group the data by year and month and count the number of orders
    orders_per_month = df.groupby(['Year', 'Month']).size().reset_index(name='Number of Orders')

    # Plot a line chart to visualize the number of orders per year and per month
    plt.figure(figsize=(10, 6))

    # Assign different colors to each year
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    years = orders_per_month['Year'].unique()

    for i, year in enumerate(years):
        year_data = orders_per_month[orders_per_month['Year'] == year]
        plt.plot(year_data['Month'], year_data['Number of Orders'], marker='o', color=colors[i])

    plt.xlabel('Month')
    plt.ylabel('Number of Orders')
    plt.title('Number of Orders per Year and Month')

    # Label x-axis with months
    months = calendar.month_abbr[1:]
    plt.xticks(range(1, 13), labels=months)

    plt.legend([str(year) for year in years])
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('orders_per_month.png')
    plt.show()

    # Generate PDF report
    pdf_file = 'orders_per_month_report.pdf'
    c = canvas.Canvas(pdf_file, pagesize=letter)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(160, 800, 'Analysis Report - Number of Orders per Month per Year')
    c.drawImage('orders_per_month.png', 50, 550, width=800, height=400)
    c.save()

    # Generate Excel file
    excel_file = 'orders_per_month_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        orders_per_month.to_excel(writer, sheet_name='Orders per Month per Year')
        
#option 8
def analyze_reviews_per_month():
    # Merge the orders and reviews dataframes
    merged_df = pd.merge(orders_2016_2020_Dataset, review_dataset, on='order_id', how='inner')

    
    # Extract year and month from the order_purchase_timestamp
    merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
    merged_df['year'] = merged_df['order_purchase_timestamp'].dt.year
    merged_df['month'] = merged_df['order_purchase_timestamp'].dt.month

    # Group the merged data by year and month, and calculate the number of orders per month
    orders_per_month_per_year = merged_df.groupby(['year', 'month']).size().reset_index(name='num_orders')

    # Convert the 'stars' column to numeric dtype
    merged_df['stars'] = pd.to_numeric(merged_df['stars'], errors='coerce')

    # Group the merged data by year and month, and calculate the average review score
    avg_review_scores_per_month_per_year = merged_df.groupby(['year', 'month'])['stars'].mean().reset_index(name='avg_review_score')

    # Get the month names
    orders_per_month_per_year['month_name'] = orders_per_month_per_year['month'].apply(lambda x: calendar.month_name[x])

    return orders_per_month_per_year, avg_review_scores_per_month_per_year



    # Plotting the number of orders and average review scores per month per year
    plt.figure(figsize=(12, 6))
    ax = sns.lineplot(data=orders_per_month_per_year, x='month_name', y='num_orders', hue='year', marker='o', label='Orders')
    ax = sns.lineplot(data=avg_review_scores_per_month_per_year, x='month_name', y='avg_review_score', hue='year', marker='o', label='Average Review Score')
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.title('Number of Orders and Average Review Scores per Month per Year')
    plt.show()

    # Generate PDF report
    pdf_file = 'reviews_for_orders_per_month_report.pdf'
    c = canvas.Canvas(pdf_file, pagesize=letter)
    c.setFont('Helvetica-Bold', 16)
    c.drawString(160, 800, 'Analysis Report - Average Review Score for Number of Orders per Month per Year')
    c.drawImage('reviews_for_orders_per_month.png', 50, 550, width=800, height=400)
    c.save()

   # Generate Excel file
    excel_file = 'reviews_for_orders_per_month_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        avg_review_scores_per_month_per_year.to_excel(writer, sheet_name='Average Review Score per Month per Year')
# Option 9: Analysis of Number of Orders Across Parts of a Day

def analyze_orders_per_day():
    # Read the orders dataset
    orders_df = pd.read_csv("C:/Users/VICTUS/Desktop/Yoshop/orders_2016-2020_Dataset.csv")

    # Convert the 'Order Date and Time Stamp' column to datetime type
    orders_df['Order Date and Time Stamp'] = pd.to_datetime(orders_df['Order Date and Time Stamp'], format="%d-%m-%Y %H:%M:%S %z")

    # Extract the date from the 'Order Date and Time Stamp' column
    orders_df['date'] = orders_df['Order Date and Time Stamp'].dt.date

    # Group the data by date and count the number of orders
    orders_per_day = orders_df.groupby('date').size().reset_index(name='num_orders')

    # Plotting the number of orders per day
    plt.figure(figsize=(12, 6))
    ax = sns.lineplot(data=orders_per_day, x='date', y='num_orders', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Number of Orders')
    plt.title('Number of Orders per Day')
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Save the plot as an image
    plot_file = 'orders_per_day_plot.png'
    plt.savefig(plot_file)
    
    # Generate PDF report
    pdf_file = 'orders_per_day_report.pdf'
    with PdfPages(pdf_file) as pdf:
        plt.savefig(pdf, format='pdf')
        plt.close()

    # Generate Excel file
    excel_file = 'orders_per_day_report.xlsx'
    with pd.ExcelWriter(excel_file) as writer:
        orders_per_day.to_excel(writer, sheet_name='Orders per Day')

    # Display the plot
    plt.show()



# Option 10: Full Report
def generate_full_report():
    analyze_reviews()
    analyze_payment_methods()
    analyze_consumer_states()
    analyze_consumer_cities()
    analyze_product_categories()
    analyze_reviews_product_categories()
    analyze_orders_per_month()
    analyze_reviews_per_month()
    analyze_orders_per_day()
    pass

# Main program
def main():
    choice = int(input("Enter the number to see the analysis of your choice: "))

    if choice == 1:
        analyze_reviews()
    elif choice == 2:
        analyze_payment_methods()
    elif choice == 3:
        analyze_consumer_states()
    elif choice == 4:
        analyze_consumer_cities()
    elif choice == 5:
        analyze_product_categories()
    elif choice == 6:
        analyze_reviews_product_categories()
    elif choice == 7:
        analyze_orders_per_month()
    elif choice == 8:
        analyze_reviews_per_month()
    elif choice == 9:
        analyze_orders_per_day()
    elif choice == 10:
        generate_full_report()
    else:
        print("Invalid choice. Please enter a valid option.")

# Run the main program
main()




```
