```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
data = pd.read_excel('Yoshops Survey_1786_Updated_13_March.xlsx')

```

```python
data
```

```python
# Extract the required columns for analysis
monthly_data = data[['Submitted Time','7. What is your Favourite food biryani essay?']]
```

```python
# Clean the data
monthly_data['Submitted Time'] = pd.to_datetime(monthly_data['Submitted Time'])
```

```python
# Group data by month for monthly report
monthly_report = monthly_data.groupby(pd.Grouper(key='Submitted Time', freq='M')).count()
```

```python
monthly_report
```

```python
# Create the bar plot for monthly report
plt.figure(figsize=(10, 6))
plt.bar(monthly_report.index.strftime('%b %Y'), monthly_report['7. What is your Favourite food biryani essay?'])
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Monthly Report: Count of Biryani Preference'
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Convert 'Submitted Time' to datetime
data['Submitted Time'] = pd.to_datetime(data['Submitted Time'], dayfirst=True)

# Group by year and count the occurrences
yearly_counts = data['Submitted Time'].dt.year.value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
yearly_counts.plot(kind='bar', color='blue')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Yearly Report')
plt.show()
```

######    We can observe the submission trends over the years based on the height of the bars. Higher bars indicate a higher number of submissions for a particular year.



```python

food_column = data['7. What is your Favourite food biryani essay?']
price_column = data['8. What is Price Rang for 1kg Biryani( 2 Person Eat) you like must ']

# Count the occurrences of each unique value
food_counts = food_column.value_counts()
price_counts = price_column.value_counts()

# Create the pie chart
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.pie(food_counts, labels=food_counts.index, autopct='%1.1f%%')
ax1.set_title('Favorite Biryani ')
ax2.pie(price_counts, labels=price_counts.index, autopct='%1.1f%%')
ax2.set_title('Biryani Price Range')

plt.tight_layout()
plt.show()
```
